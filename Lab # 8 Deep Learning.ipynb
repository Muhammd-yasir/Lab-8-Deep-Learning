1. Loading an Image

```python
import cv2
import matplotlib.pyplot as plt
# Load an image from a file
image = cv2.imread('image.jfif')
# Convert the image from BGR to RGB (OpenCV loads images in BGR format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Display the image
plt.imshow(image_rgb)
plt.title('Loaded Image')
plt.axis('off') # Hide axis
plt.show()
```

```python
# Resize the image to 200x200 pixels
resized_image = cv2.resize(image_rgb, (200, 200))
# Display the resized image
plt.imshow(resized_image)
plt.title('Resized Image (200x200)')
plt.axis('off')
plt.show()
```

```python
# Apply horizontal flip
flipped_image = cv2.flip(image_rgb, 1)
# Display the flipped image
plt.imshow(flipped_image)
plt.title('Horizontally Flipped Image')
plt.axis('off')
plt.show()
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
# Use Canny edge detection
edges = cv2.Canny(gray_image, 100, 200)
# Display the edges
plt.imshow(edges, cmap='gray')
plt.title('Edges in Image')
plt.axis('off')
plt.show()
```

```python
# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# Display the histogram for the grayscale image
plt.hist(gray_image.ravel(), bins=256, range=[0,256])
plt.title('Histogram of Grayscale Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
# Display the histogram for each color channel (RGB)
colors = ('r', 'g', 'b')
for i, color in enumerate(colors):
 histogram = cv2.calcHist([image_rgb], [i], None, [256], [0, 256])
 plt.plot(histogram, color=color)
 plt.xlim([0, 256])
plt.title('Histogram of Color Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load an image from a file
image = cv2.imread('image.jfif')

# Convert the image from BGR to RGB (OpenCV loads images in BGR format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert RGB image to HSV
image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)

# Increase the brightness (Value channel)
brightness_factor = 50  # Adjust this value as needed
image_hsv[:,:,2] += brightness_factor
image_hsv[:,:,2] = np.clip(image_hsv[:,:,2], 0, 255)  # Clip values to 0-255 range

# Convert back to RGB
brightened_image_rgb = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB)

# Display the original and brightened images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(brightened_image_rgb)
plt.title('Brightened Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image from a file
image = cv2.imread('image.jfif')

# Convert the image from BGR to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Sobel edge detection
sobel_x = cv2.Sobel(image_gray, cv2.CV_64F, 1, 0, ksize=3)  # Sobel edge detection along x-axis
sobel_y = cv2.Sobel(image_gray, cv2.CV_64F, 0, 1, ksize=3)  # Sobel edge detection along y-axis

# Combine the results from x-axis and y-axis
sobel_combined = np.sqrt(sobel_x**2 + sobel_y**2)

# Normalize the result to the range [0, 255]
sobel_combined = cv2.normalize(sobel_combined, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)

# Display the result
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image from a file
image = cv2.imread('image.jfif')

# Convert the image from BGR to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Sobel edge detection
sobel_x = cv2.Sobel(image_gray, cv2.CV_64F, 1, 0, ksize=3)  # Sobel edge detection along x-axis
sobel_y = cv2.Sobel(image_gray, cv2.CV_64F, 0, 1, ksize=3)  # Sobel edge detection along y-axis

# Combine the results from x-axis and y-axis
sobel_combined = np.sqrt(sobel_x**2 + sobel_y**2)

# Normalize the result to the range [0, 255]
sobel_combined = cv2.normalize(sobel_combined, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)

# Display the images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load an image from a file
image = cv2.imread('image.jfif')

# Convert the image from BGR to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Sobel edge detection along x-axis
sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)

# Apply Sobel edge detection along y-axis
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)

# Combine the results from x-axis and y-axis
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# Display the images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')

plt.show()
```
